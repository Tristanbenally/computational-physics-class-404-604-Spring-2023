```python
#Write a program to numerically evaluate the derivative of f(x)=cos x at an arbitrary position x.

#Evaluate the derivative at position x using the forward difference scheme. 
#Plot log(Error)=log10|f’(numerically)-f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at x=π/4.
#You should find Error~dx^n
#what is n? what is the optimal value for dx.

#Repeat the exercise using the central difference scheme. Why this scheme is better?

#Repeat (2) for x=0. Why do the results differ? 
#(hint: write out the error term derived in the class explicitly by plugging the derivative of cos or sin)
import numpy as np
#forward difference scheme with f(x) = cosx with x=pi/4
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

x=np.pi/4
dx=0.1
print('forward difference',derifw(np.cos,x,dx))
print('analytical result',np.sin(-x))
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16, -1, 0.1):
    y=derifw(np.cos,x,dx)
    print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
    
    
import matplotlib.pyplot as plt

plt.plot(dxarr, abs(yarr-np.sin(-x)), 'b-', lw=2)     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()

#Error dx^n = 10^-8
```

```python
import numpy as np
import matplotlib.pyplot as plt
#central difference scheme with f(x)=cosx at x = pi/4
def dericd(y,x,dx): 
    return(y(x+dx/2)-y(x-dx/2))/dx
x=np.pi/4
dx=0.1
print('central difference',dericd(np.cos,x,dx))
print('analytical result',np.sin(-x))
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16, -1, 0.1):
    y=dericd(np.cos,x,dx)
    print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
    
    
import matplotlib.pyplot as plt

plt.plot(dxarr, abs(yarr-np.sin(-x)), 'r-', lw=2)     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()

#central scheme is better due to being at dx = 10^-8 the scheme has a lower Error difference
```
