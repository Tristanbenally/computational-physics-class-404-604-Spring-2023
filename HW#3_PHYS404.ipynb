```python
import numpy as np
import matplotlib.pyplot as plt

#Problem 1 using memorization method to calculate F(n)/F(n-1) with plot

# if we have defined an array table = [None]*(n+1)

def fib3(n, table): # Store the data we already calculated, linear space and time complexity, but still have the stack limit
    if n == 0 or n==1:
        table[n]=n
    
    # if the value has not been calculated, calculate it 
    if table[n] is None:
        table[n] = fib3(n-1,table) + fib3(n-2,table) # try this:  please finish this line using recursive method
        
    return table[n]

# calclate time with n= as specified

import time
n=50 #**
ratio=[]
for n in range(2,n+1): #since F(0)=0 for F(n-1) so start at 2
    table = [None]*(n+1)
    start_time = time.time()
    #print(fib3(n, table))
    #print(fib3(n-1, table))
    y = fib3(n,table)/fib3(n-1,table)
    ratio.append(y)
print("--- %s seconds ---" % (time.time() - start_time))
print(ratio)
x = np.arange(2,n+1)
#print(y)
print(x)
plt.plot(x,ratio,color='green')
plt.show()

# In programming, the call stack is a buffer that stores requests that need to be handled.
```

```python
#Problem 2 plotting n = 1,10,100,1000,10000,...
#we will use recursive method up for cases n=1,10,100
#we will use tabulation method for cases n=1000
#we will use matrix method for cases n=10000
# Recursion method fib2, 

def fib2(n):      # extremely simple, but exponentially running time, very slow for large n, n cannot too large (stack limit)
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib2(n-1)+fib2(n-2)
    
import time
#time sets for recursive method
time1 = []
time2 = []
time3 = []
#for n=1
start_time = time.time()
x = print(fib2(1))
t = print("--- %s seconds ---" % (time.time() - start_time))
time1.append(t)
#for n=10
for n in range(2,11):
    x_phi = print(fib2(10))
    t_phi = print("--- %s seconds ---" % (time.time() - start_time))
    time2.append(t_phi)
#for n=100
for n in range(2,101):
    x_phi2 = print(fib2(100))
    t_phi2 = print("--- %s seconds ---" % (time.time() - start_time))
    time3.append(t_phi2)


n1 = np.arange(1)
n2 = np.arange(2,11)
n3 = np.arange(2,101)

#log scale for t

#plot in log
plt.title('')
plt.xlabel("n")
plt.ylabel("log(t)")
plt.plot(n1,np.log(time1),color ='green')
plt.plot(n2,np.log(time2),color = 'red')
plt.plot(n3,np.log(time3),color = 'purple')
plt.show()

#Tabulation method fib4,
# calculate numbers one by one from table[2] to table[n]

def fib4(n):
    table=[0]*(n+1)
    table[1] = 1
       
    # try this: please finish the following two lines calcuating table[i] from table[i-1] and table[i-2]
    for n in range(2,n+1,1):
        table[n]= table[n-1] + table[n-2]
        
    return table[n]

time5 = []

import time
start_time = time.time()
#for n=1000
for n in range(2,1001):
    x_phi4 = print(fib4(1000))
    t_phi4 = print("--- %s seconds ---" % (time.time() - start_time))
    time5.append(t_phi4)

n4 = np.arange(2,1001)

#plot in log
plt.title('')
plt.xlabel("n")
plt.ylabel("log(t)")
plt.plot(n4,np.log(time5),color ='orange')
plt.show()

#Matrix method fib5,

def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib5(n):
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f

import time
start_time = time.time()
#for n=10000
time6 = []
for n in range(2,10001):
    x_phi5 = print(fib5(10000))
    t_phi5 = print("--- %s seconds ---" % (time.time() - start_time))
    time6.append(t_phi5)
    
n5 = np.arange(2,100001)

#plot in log
plt.title('')
plt.xlabel("n")
plt.ylabel("log(t)")
plt.plot(n5,np.log(time6),color ='blue')
plt.show()
```
